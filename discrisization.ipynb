```python
import pandas as pd

import numpy as np
```

```python
data=pd.read_excel('heartP.xlsx')
data
```

```python
del data['Unnamed: 0']
```

```python
age=data['age']

ageCat = []


for i in range(len(age)):
    if (age[i]>=25 and age[i]<=45):
        ageCat.append('young')
        
    elif (age[i]>45 and age[i]<=65):
        ageCat.append('middle')
    
    elif(age[i]>65):
        ageCat.append('old')
        
ageCat        
```

```python
data['age'] = ageCat

data
```

```python
from sklearn.preprocessing import KBinsDiscretizer

kbins = KBinsDiscretizer(n_bins=3,encode='ordinal',strategy='kmeans')

li=['bp','chol','hr']

data[li]=kbins.fit_transform(data[li])

data
```
